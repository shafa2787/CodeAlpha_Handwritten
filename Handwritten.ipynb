```python

# Read the first image file
img1 = mpimg.imread("/content/drive/MyDrive/CodeAlpha/2.Hand_written/handwritten/Img/img009-034.png")

# Read the second image file
img2 = mpimg.imread("/content/drive/MyDrive/CodeAlpha/2.Hand_written/handwritten/Img/img001-014.png")

# Create a figure and axes
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Plot the first image
axes[0].imshow(img1)
axes[0].set_title('Image 1')

# Plot the second image
axes[1].imshow(img2)
axes[1].set_title('Image 2')

# Remove the axis ticks
for ax in axes:
    ax.axis('off')

# Show the plot
plt.show()
```
